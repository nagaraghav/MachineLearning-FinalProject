```python
import pandas as pd
import holidays
```

```python
dataset = pd.read_csv('./dataset/weatherwithtripcount.csv')
```

```python
dataset['events'].unique()
```

```python
def isRain(r):
    e_c = str(r['events'])
    if 'Rain' in e_c or 'rain' in e_c:
        return 1
    return 0

def isFog(r):
    e_c = str(r['events'])
    if 'Fog' in e_c:
        return 1
    return 0

def isThunderstorm(r):
    e_c = str(r['events'])
    if 'Thunderstorm' in e_c:
        return 1
    return 0
```

```python
dataset['isRain'] = dataset.apply(isRain,axis=1)
dataset['isFog'] = dataset.apply(isFog,axis=1)
dataset['isThunderstorm'] = dataset.apply(isThunderstorm,axis=1)
```

```python
dataset[dataset['events'].notna()]
```

```python
dataset.drop(['Unnamed: 0', 'precipitation_inches', 'events', 'wind_dir_degrees'], axis=1,inplace=True)
```

```python
dataset.columns
```

```python
import datetime
weekdays=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
def weekday(r):
    d = r['date']
    i = datetime.datetime.strptime(d,"%Y-%m-%d").weekday()
    return weekdays[i]
```

```python
dataset['Weekday'] = dataset.apply(weekday, axis=1)
```

```python
us_holidays=holidays.US()
def holiday(r):
    d = r['date']
    h = us_holidays.get(d)
    if h is not None:
        return h
    else:
        return 'None'
```

```python
dataset['Holiday'] = dataset.apply(holiday, axis=1)
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,columns=['zip_code', 'Weekday', 'Holiday'])
```

```python
dataset_nodate = dataset.drop(['date'], axis=1)
```
