```python
us_holidays=holidays.US()
```

```python
def add_holiday(dataset):
    holiday=[]
    for d in dataset["date"]:
        if(us_holidays.get(d)!=None):
            holiday.append(us_holidays.get(d))
        else:
            holiday.append("none")
    dataset["holiday"]=holiday
```

```python
add_holiday(train_weather)
```

```python
def add_weekdays(dataset):
    weekdays=[]
    week=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
    for d in dataset["date"]:
        day=datetime.datetime.strptime(d,"%m/%d/%Y").weekday()
        weekdays.append(week[day])
    dataset["weekdays"]=weekdays
```

```python
add_weekdays(train_weather)
```

```python
def create_tripAmount_df(train_trip):
    data=[]
    date_save=train_trip['date'][0]
    count=0
    for trip_date in train_trip['date']:
        if(trip_date==date_save):
            count=count+1
        else:
            data_day=[date_save,count]
            data.append(data_day)
            count=1
            date_save=trip_date
    return pd.DataFrame(data,columns=['date','trip_amount'])
```
