```python
import pandas as pd
```

```python
weather = pd.read_csv('./dataset/weather.csv')
print(weather.shape)
print(weather.columns)
trip = pd.read_csv('./dataset/groupedByZip_count.csv')
print(trip.shape)
print(trip.columns)
```

```python
weather.head(10)
```

```python
weather['date'] = pd.to_datetime(weather['date'])
weather['date'] = weather['date'].dt.date
zip_weather = weather.groupby('zip_code')
```

```python
weather
```

```python
def selectDateZip(row):
    qzip = row['zip_code']
    qdate = str(row['date'])
    #print(str(qdate),type(qzip))
    res = trip.loc[(trip['city_zip'] == qzip) & (trip['date'] == qdate)]['trip_count'].to_numpy()

    if res.shape == (0,):
        return 0
    return res[0]
```

```python
weather['trip_count'] = weather.apply(selectDateZip,axis=1)
```
